## Data cleaning

```python
#Dropping unnecessary columns
df.drop('atemp',axis=1,inplace=True)
df.drop('holiday',axis=1,inplace=True)
df.drop('dteday',axis=1,inplace=True)
df.drop('instant',axis=1,inplace=True)
```

```python

df.head()
```

```python
#Dealing with Categorical varibales 'season' & 'weathersit'
df['season'].replace({1:"spring",2:"summer",3:"fall",4:"winter"},inplace=True)
df['weathersit'].replace({1:"Clear_Few clouds",2:"Mist_Cloudy",3:"Light Snow_Rain",4:"HeavyRain"},inplace=True)
df['mnth'].replace({1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'June',7:'July',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'},inplace=True)
df['weekday'].replace({0:'mon',1:'Tues',2:'Wed',3:'Thurs',4:'Fri',5:'Sat',6:'Sun'},inplace=True)
```

```python
df.weekday.value_counts()
```

```python
df.head()
```

```python
df.info()
```

```python
#Convert variables to object type
df['mnth']=df['mnth'].astype(object)
df['season']=df['season'].astype(object)
df['weathersit']=df['weathersit'].astype(object)
df['weekday']=df['weekday'].astype(object)
```

```python
#Converting categorical variables to dummy variables

df = pd.get_dummies(df,columns=['season'],drop_first=True)
df = pd.get_dummies(df,columns=['weathersit'],drop_first=True)
df = pd.get_dummies(df,columns= ['mnth'],drop_first=True)
df = pd.get_dummies(df,columns = ['weekday'],drop_first=True)
```

```python
df.head()
```

```python
df.head()
```

```python
#Dropping variables 'casual' & 'registered' as we already have cnt as addition of both
df.drop('casual',axis=1,inplace=True)
df.drop('registered',axis=1,inplace=True)
df.head()
```

```python
df.info()
```

```python
#Split data into train & test data
from sklearn.model_selection import train_test_split
np.random.seed(0)
df_train,df_test = train_test_split(df,train_size = 0.7,test_size=0.3 ,random_state = 100)
```
